```python
import pandas as pd
import json
```

```python
df = pd.read_csv("./dataTest_device_profile.csv")
df['params'] = df['params'].apply(lambda x: json.loads(x))
df.head()
```

```python
formatted = pd.json_normalize(df['params'])
```

```python
new_df = pd.concat([df[['timestamp', 'userId']], formatted], axis=1)
new_df.head()
```

```python
new_df.groupby(['userId']).agg({'hw_device_id': 'count'}).sort_values(by='hw_device_id', ascending=False)
```

```python
user_list = new_df['userId'].unique()
user_list
```

```python
device_list = new_df['hw_device_id'].unique()
device_list
```

```python
new_df.to_csv('./data_device_profile_formatted.csv', index=False)
```
